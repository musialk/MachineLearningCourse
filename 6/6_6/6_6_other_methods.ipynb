# 6.6. Pozostałe metody klastrowania

K-Means to oczywiście nie jedyna metoda klastrowania. Co więcej zakłada ona, że znamy od samego początku liczbę grup. Omówimy jedną z metod, która jest w stanie sama dopasować również ten parametr, dzięki czemu możemy wykryć nie tylko regularności, których obecność zakładamy.

## DBSCAN

Metoda ta poszukuje obserwacji, które mają w swoim sąsiedztwie jak najwięcej innych przykładów. Zakłada ona, że dane są zbudowane z obszarów o wysokim oraz niskim zagęszczeniu danych i stara się odnaleźć takie miejsca, które skupiają jak najwięcej wektorów danych. Parametrami tej metody są m.in. odległość graniczna, do której powinno rozpatrywać się sąsiedztwo punktów oraz minimalna licza obserwacji, która może tworzyć pojedynczy klaster.

Powróćmy do przykładu, z którym KMeans nie poradził sobie zbyt dobrze i zobaczmy co na jego temat będzie w stanie odnaleźć metoda DBSCAN.

```python
from sklearn.datasets import make_circles
```

```python
X_circle, labels_circle = make_circles(n_samples=1000, noise=0.05,
                                       factor=0.5, random_state=326321)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(12.5, 12.5))
plt.scatter(x=X_circle[:, 0], y=X_circle[:, 1], c=labels_circle, 
            cmap="Set1", edgecolor=(0, 0, 0, 1))
```

```python
from sklearn.cluster import DBSCAN
```

```python
dbscan = DBSCAN(eps=0.2)
predicted_labels = dbscan.fit_predict(X_circle)
```

```python
plt.figure(figsize=(12.5, 12.5))
plt.scatter(x=X_circle[:, 0], y=X_circle[:, 1], c=predicted_labels, 
            cmap="Set1", edgecolor=(0, 0, 0, 1))
```

DBSCAN nie tworzy metody, która pozwala na określenie przynależności dla nowych punktów, a jedynie pozwala pogrupować dane, które już posiadamy.
