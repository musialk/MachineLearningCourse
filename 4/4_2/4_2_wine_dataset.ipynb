## Łączenie zbiorów danych

Jak widać powyżej, oba zbiory mają tę samą strukturę, a więc ich połączenie zdaje się być banalnie proste. Jest jednak jedna dodatkowa informacja, która zakodowana została w nazwie pliku - jest to kolor wina. Dodamy tę informację do zbioru, ponieważ może to być dość istotne.

```python
wine_quality_df = pd.concat([
    pd.read_csv("./data/wine-quality/winequality-white.csv", 
                delimiter=";")
        .assign(color="white"),
    pd.read_csv("./data/wine-quality/winequality-red.csv", 
                delimiter=";")
        .assign(color="red"),
])
wine_quality_df.sample(n=5)
```

## Podstawowe zrozumienie dostarczonych danych

Zbiór, który otrzymaliśmy, jest niekoniecznie jasny dla tych, którzy nie są koneserami wina. Jest on zbudowany z szeregu atrybutów fizykochemicznych i posiadanie jakiejkolwiek bazowej intuicji może być trudne. W związku z tym przyjrzymy się kolejnym atrybutom, żeby zyskać jakąkolwiek wiedzę na temat tego jak wygląda zbiór, który będziemy modelować. Najpierw, warto jest zazwyczaj sprawdzić jak kształtuje się korelacja, aby wychwycić które zmienne są z sobą najmocniej związane.

```python
wine_quality_df.corr() \
    .style.background_gradient(cmap="coolwarm")
```

```python
wine_quality_df.describe().T
```

```python
wine_quality_df.plot.box(figsize=(25.0, 7.5), 
                         rot=90)
```

Z powyższego wykresu można wysnuć kilka wniosków. Jednym z nich jest z pewnością fakt, iż prezentacja cech o wartościach pochodzących z różnych zakresów dość mocno utrudnia wyłapanie prawidłowości w zbiorze danych.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
numeric_columns_df = wine_quality_df \
    .select_dtypes(include=np.number)
```

```python
fig, ax = plt.subplots(6, 2, figsize=(16.0, 32.0))
flat_ax = ax.flatten()
for idx, column_name in enumerate(numeric_columns_df.columns):
    wine_quality_df[column_name].plot.box(ax=flat_ax[idx])
```

## Opis atrybutów zbioru danych

Skupmy się jednak na stworzeniu słownika, który wytłumaczy nam każdą pojedynczą cechę, abyśmy choć podstawowo rozumieli co kryje się za poszczególnymi wartościami. Tego typu słownik nazywa się często ***Business Glossary*** i na pewno spotkacie się z tym pojęciem jeszcze wiele razy.

* **fixed acidity** - zawartość [kwasu winowego](https://pl.wikipedia.org/wiki/Kwas_winowy), w g/L
* **volatile acidity** - zawartość [kwasu octowego](https://pl.wikipedia.org/wiki/Kwas_octowy), w g/L
* **citric acid** - zawartość kwasu cytrynowego, w g/L
* **residual sugar** - zawartość cukru resztkowego
* **chloride** - zawartość chlorku sodu, w g/L
* **free sulfur dioxide** - zawartość *wolnego* dwutlenku siarki, który wykazuje wciąż działanie konserwujące, w mg/L
* **total sulfur dioxide** - łączna zawartość *wolnego* i *związanego* dwutlenku siarki, w mg/L
* **density** - gęstość, w g/cm^3
* **pH** - ilościowa skala kwasowości/zasadowości
* **sulphates** - zawartość siarczyn, w g/L
* **alcohol** - zawartość alkoholu w procentach
* **quality** - jakość wina, ocena ekspercka, liczba naturalna z przedziału \[0,10\]
* **color** - kolor wina, biały lub czerwony
